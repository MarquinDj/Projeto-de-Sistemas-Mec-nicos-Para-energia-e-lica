## **IMPORTANDO BIBLIOTECAS**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

## **TRATANDO OS DADOS**

```python
dados = pd.read_csv('ITABAIANINHA - windData.csv', on_bad_lines='skip', sep=';')
```

```python
dados.head(20)
```

```python
dados.isnull().sum()
```

```python
columns = ['Temperatura', 'Umidade', 'Pressao', 'Vento', 'Nebulosidade', 'Insolacao', 'TemperaturaMaxima', 'TemperaturaMinima', 'Chuva']

for column in columns:
    dados[column].fillna(dados[column].mean(), inplace=True)

dados['Velocidade do vento'].dropna(inplace=True)
```

```python
dados.isnull().sum()
```

```python
sns.boxplot(data=dados['Velocidade do vento'], orient = "h")
```

```python
dados = dados[dados['Velocidade do vento'] < 3]
```

```python
sns.boxplot(data=dados['Velocidade do vento'], orient = "h")
```

```python
dados['Data'] = pd.to_datetime(dados['Data'])
```



## **TAREFA 3**

**PARTE 1**

```python
data1 = pd.DataFrame(dados["Data"].iloc[0::95])
vel1 = pd.DataFrame(dados["Velocidade do vento"].iloc[0::95])
tarefa1 = pd.concat([data1, vel1], axis=1)
tarefa1
```

```python
Grafico1 = sns.lineplot(data=tarefa1, x='Data', y='Velocidade do vento', color = 'red') 
Grafico1.set_title("Média da velocidade do vento por ano - Itabaianinha, SE", fontsize = 20)
Grafico1.set_xlabel("Data", fontsize = 20)
Grafico1.set_ylabel("Velocidade do vento (m/s)", fontsize = 20)
Grafico1.figure.set_size_inches(26, 8)
Grafico1.set_ylim(0, 6)
```

**PARTE 2**

```python
dadosPt2 = pd.read_csv('windData_cidades_2018.csv', on_bad_lines='skip', sep=';')
```

```python
dadosPt2.info()
```

```python
dadosPt2['Data'] = pd.to_datetime(dadosPt2['Data'])
```

```python
dadosPt2.head(10)
```

```python
sns.boxplot(data=dadosPt2['Velocidade - Itabaianinha'], orient = "h")
```

```python
cols_velocidade = [col for col in dadosPt2.columns if 'Velocidade -' in col]
dadosPt2 = dadosPt2.loc[(dadosPt2[cols_velocidade] > 0).all(axis=1)]
```

```python
dadosPt2 = dadosPt2.reset_index()
dadosPt2 = dadosPt2.set_index('Data')
```

```python
dadosPt2 = dadosPt2.drop(columns=['index'])
dadosPt2.head()
```

```python
ita = pd.DataFrame(dadosPt2['Velocidade - Itabaianinha'].iloc[0::30])
cara = pd.DataFrame(dadosPt2['Velocidade - Caracarai'].iloc[0::30])
palmas = pd.DataFrame(dadosPt2['Velocidade - Palmas'].iloc[0::30])
tarefa2 = pd.concat([ita, cara, palmas], axis=1)
tarefa2
```

```python
Grafico2 = sns.lineplot(data=tarefa2)
Grafico2.figure.set_size_inches(26, 8)
Grafico2.set_title("Média da velocidade do vento nas 3 localidades / 2018 - 2019", fontdict = {'fontsize': 20})
Grafico2.set_xlabel("Tempo decorrido", fontsize = 20)
Grafico2.set_ylabel("Velocidade do vento (m/s)", fontsize = 20)
Grafico2.set_ylim(0, 10)
plt.show()
```

```python
ita2 = pd.DataFrame(dadosPt3['Velocidade - Itabaianinha'].iloc[0::30])
cara2 = pd.DataFrame(dadosPt3['Velocidade - Caracarai'].iloc[0::30])
palmas2 = pd.DataFrame(dadosPt3['Velocidade - Palmas'].iloc[0::30])
tarefa3 = pd.concat([ita2, cara2, palmas2], axis=1)
tarefa3
```

```python
Grafico3 = sns.lineplot(data=tarefa3)
Grafico3.figure.set_size_inches(26, 8)
Grafico3.set_title("Média da velocidade do vento nas 3 localidades / 2020 - 2021", fontdict = {'fontsize': 20})
Grafico3.set_xlabel("Tempo decorrido", fontsize = 20)
Grafico3.set_ylabel("Velocidade do vento (m/s)", fontsize = 20)
Grafico3.set_ylim(0, 10)
plt.show()
```
